```python
import numpy as np
import pandas as pd
import warnings

warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("/content/drive/MyDrive/RTA Dataset.csv")
df.head()
```

### Checking for NULL/Missing values so that we can manage them


```python
df.isnull().sum()
```

##### **We can see that following columns have very large count of missing values so dropping those columns.**

```python
df.drop(columns = ['Defect_of_vehicle', 'Work_of_casuality', 'Fitness_of_casuality','Service_year_of_vehicle'], inplace=True)
```

```python
df.isnull().sum()
```

##### **Creating a new column 'Hour' from Time column.**

```python
df['Time'] = pd.to_datetime(df['Time'])
df['Hour'] = df['Time'].dt.hour
df.drop('Time', axis=1, inplace=True)
```

```python
df.head()
```

```python
columns_to_fill = ['Type_of_vehicle', 'Types_of_Junction', 'Driving_experience', 'Lanes_or_Medians', 'Area_accident_occured']

for column in columns_to_fill:
    df[column].fillna(df[column].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
columns_to_fill_2 = ['Educational_level', 'Vehicle_driver_relation' ,'Owner_of_vehicle','Road_allignment','Road_surface_type', 'Type_of_collision','Vehicle_movement']
for column in columns_to_fill_2:
    df[column].fillna('Unknown', inplace=True)
```

```python
df.isnull().sum()
```

#### **Now as you can see we have Imputed values into our dataset successfully**

## One hot Encoding out features

```python
df.info()
```

```python
categorical_features = ['Day_of_week', 'Age_band_of_driver', 'Sex_of_driver',
       'Educational_level','Vehicle_driver_relation', 'Driving_experience', 'Type_of_vehicle',
       'Owner_of_vehicle', 'Area_accident_occured', 'Lanes_or_Medians',
       'Road_allignment', 'Types_of_Junction', 'Road_surface_type',
       'Road_surface_conditions', 'Light_conditions', 'Weather_conditions',
       'Type_of_collision', 'Vehicle_movement', 'Casualty_class',
       'Sex_of_casualty', 'Age_band_of_casualty', 'Casualty_severity',
       'Pedestrian_movement', 'Cause_of_accident',]
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
enc_df = pd.DataFrame(enc.fit_transform(df[categorical_features]).toarray())
# merge with main df bridge_df on key values
main_df = df.join(enc_df)
main_df
```

```python
main_df = main_df.drop(categorical_features, axis = 1)
```

```python
main_df
```

Label Encode out Target Variable

```python

from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
# Assigning numerical values and storing in another column
main_df['Accident_severity'] = labelencoder.fit_transform(main_df['Accident_severity'])
main_df
```

**Now as we know from out Exploratory Data Analysis, the dataset is highly imbalanced, so we will treat our dataset with 'SMOTENC' technique**

#Handling imbalanced Data

Before that we will define our Dependent(y) and Independent Variables(X)

```python
X = main_df.drop('Accident_severity', axis = 1)
X.head()
```

```python
y = main_df['Accident_severity']
y.head()
```

###Now using SMOTETomek

```python
X_main = X.copy()
y_main = y.copy()
```

```python
# importing the SMOTETomek object from imblearn library
from imblearn.over_sampling import SMOTENC
from imblearn.combine import SMOTETomek
sm = SMOTETomek(random_state=101, n_jobs=6)
X_Smote, y_Smote = sm._fit_resample(X_main, y_main)

X_Smote.shape, y_Smote.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_Smote, y_Smote, test_size=0.2, random_state=42)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def evaluate_classification_model(true, predicted):
    accuracy = accuracy_score(true, predicted)
    precision = precision_score(true, predicted, average='weighted')
    recall = recall_score(true, predicted, average='weighted')
    f1 = f1_score(true, predicted, average='weighted')

    return accuracy, precision, recall, f1
```

```python
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

models = {
    "XGBoost Classifier": XGBClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    #"Support Vector Machine": SVC(),
    "Gradient Boosting": GradientBoostingClassifier()
}

model_list = []
accuracy_list = []

for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(X_train, y_train)

    # Make predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    # Evaluate Train and Test dataset
    accuracy, precision, recall, f1 = evaluate_classification_model(y_test, y_test_pred)

    print(list(models.keys())[i])
    model_list.append(list(models.keys())[i])

    print('Model performance for Test set')
    print("- Accuracy: {:.4f}".format(accuracy))
    print("- Precision: {:.4f}".format(precision))
    print("- Recall: {:.4f}".format(recall))
    print("- F1 Score: {:.4f}".format(f1))
    accuracy_list.append(accuracy)

    print('='*35)
    print('\n')
```

```python
pd.DataFrame(list(zip(model_list, accuracy_list)), columns=['Model Name', 'Accuracy']).sort_values(by=["Accuracy"],ascending=False)
```

**As from above analysis, we can say Random Forest provides us with best Accuracy. Hence, we will do hyperparamter tuning for RandomForestClassifier and train our Fianl Model**

#Grid Search CV

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [200, 500, 800, 1000],
    'max_depth' : [6,12,18,24],
    'criterion' :['gini', 'entropy']
}

rfc=RandomForestClassifier(random_state=42)
CV_rfc = GridSearchCV(estimator=rfc, param_grid=param_grid, cv= 5)
CV_rfc.fit(X_train, y_train)
```

```python
CV_rfc.best_params_
```

#**Building our Main Model**

```python
main_model = RandomForestClassifier(n_estimators=200, max_depth=24, criterion ='gini' ,random_state = 42)
main_model.fit(X_train,y_train)
```

```python
y_preds_main_model = main_model.predict(X_test)
accuracy, precision, recall, f1 = evaluate_classification_model(y_test, y_preds_main_model)
```

```python
print(f"Score {main_model.score(X_train, y_train)}")
print("- Accuracy: {:.4f}".format(accuracy))
print("- Precision: {:.4f}".format(precision))
print("- Recall: {:.4f}".format(recall))
print("- F1 Score: {:.4f}".format(f1))

```

```python
from sklearn.metrics import classification_report
classification_report = classification_report(y_test,y_preds_main_model)
print(classification_report)
```
